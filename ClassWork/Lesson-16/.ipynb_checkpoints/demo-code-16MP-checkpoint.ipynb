## Exploring Rossmann Drug Store Sales Data

# Lesson 16

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv('dataset/rossmann.csv', skipinitialspace=True, low_memory=False)
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
# Filter to Store 1
store1_data = data[data.Store == 1]
```

```python
# Filter to open days
store1_open_data = store1_data[store1_data.Open==1]
```

```python
# Plot the sales over time
store1_open_data[['Sales']].plot()
```

```python
print store1_data.Sales.autocorr(lag=1) # -0.12
print store1_data.Sales.autocorr(lag=2) # -0.03
```

```python
from pandas.tools.plotting import autocorrelation_plot
autocorrelation_plot(store1_data.Sales)
print("This shows that as the lag gets bigger there is less correlation")
print("And we don't know which number to use")
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(store1_data.Sales, lags=100)
print("This can show us where the most correlation is, which is at 7 days, or the previous week")
```

##We can expand the window to 25 days to see that the random spikes occur regularly at 7 days.  What does this mean?


```python
print plt.figure(num=1)
plot_acf(store1_data.Sales, lags=25)
```

```python
from statsmodels.tsa.arima_model import ARMA
#Need to store things as float to make sure that we're getting full length numbers.
store1_sales_data = store1_open_data[['Sales']].astype(float)
model = ARMA(store1_sales_data, (1, 0)).fit()
model.summary()
```

```python
model = ARMA(store1_sales_data, (2, 0)).fit()
model.summary()
```

```python
model.resid.plot()
```

```python
plot_acf(model.resid, lags=50)
```

```python
model = ARMA(store1_sales_data, (1, 1)).fit()
model.summary()
```

```python
#This is a literal addition of the two models. Our coefficients here show us that that the Autocorrelation
#Has a greater impact and that it brings things down and it's a decreasing model. Also, the MA value can't be
#read in the same fashion as a coefficient because it's multiplying an error that could be positive or negative
#and then the change could be different'''
```

#ARIMA

```python
#Remember that the second number in an Arima is the difference (P,D,Q) or previous terms, differences, errors
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(store1_sales_data, (1, 0, 1)).fit()
model.summary()
```

```python
model = ARIMA(store1_sales_data, (1, 1, 1)).fit()
model.summary()
```

```python
model = ARIMA(store1_sales_data, (1, 1, 0)).fit()
model.summary()
```

```python
store1_sales_data.Sales.diff(1).autocorr(1) #-0.181
```

```python
store1_sales_data.Sales.diff(1).plot()
```

##Predictions

```python
fig, ax = plt.subplots()
store1_sales_data['2014'].plot(ax=ax)
model.plot_predict(0, 200, ax=ax, plot_insample=False)


```

### Fixing the holiday issue

```python
#Here we go to more data points, and we can see we've gotten rid of the seasonal autocorrelation.
model = ARIMA(store1_sales_data, (7, 1, 2)).fit()
model.summary()
#plot_acf(model.resid, lags=50)
fig, (ax1, ax2) = plt.subplots(nrows=1,ncols=2)
model.plot_predict(0, 200, ax=ax1, plot_insample=False)
```
